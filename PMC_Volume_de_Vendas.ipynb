```python
!pip install sidrapy

import pandas as pd
import sidrapy as sidra
```

```python
dados_8880 = pd.read_json(path_or_buf = "https://apisidra.ibge.gov.br/values/t/8880/n1/all/v/all/p/all/c11046/56734/d/v7169%205,v7170%205,v11708%201,v11709%201,v11710%201,v11711%201?formato=json")
dados_8880
```

```python
dados_limpos = dados_8880[["V", "MN", "D2N", "D3C"]]
dados_limpos = dados_limpos.drop(index=0)
dados_limpos = dados_limpos.rename(columns={"V": "Valor", "MN": "Unidade de Medida", "D2N": "Variável", "D3C": "Data"})
dados_limpos = dados_limpos[["Data", "Unidade de Medida", "Variável", "Valor"]]
#os dados já estão em termos mensais #2022 = 100 (base)
dados_limpos['Data'] = dados_limpos['Data'].str.split(' ').str[0]
dados_limpos['Data'] = pd.to_datetime(dados_limpos['Data'], format='%Y%m')
dados_limpos['Data'] = dados_limpos['Data'].dt.strftime('%Y-%m')
dados_limpos['Valor'] = dados_limpos['Valor'].replace('...', pd.NA) #Substituição por NaN: As reticências geralmente representam valores ausentes ou inválidos em conjuntos de dados. Substituí-las por NaN permite que o Pandas as reconheça como valores ausentes para posterior manipulação.
dados_limpos = dados_limpos.dropna(subset=['Valor']) #Remoção de linhas com NaN: O método dropna com subset=['Núm_Índice'] remove as linhas do DataFrame onde a coluna "Núm_Índice" contém NaN.
dados_limpos['Valor'] = pd.to_numeric(dados_limpos['Valor']) #Converta a coluna "Núm_Índice" para o tipo numérico, caso necessário:
dados_limpos
```

```python
dados_limpos = dados_limpos.drop(columns=['Unidade de Medida'])
unique_unidades = dados_limpos["Variável"].unique()
print(unique_unidades)
```

```python
dados_limpos["Variável"] = dados_limpos["Variável"].replace("PMC - Número-índice (2022=100)", "Nº Índice (base 2022)")
dados_limpos["Variável"] = dados_limpos["Variável"].replace("PMC - Número-índice com ajuste sazonal (2022=100)", "Nº Índice_A.S (base 2022)")
dados_limpos["Variável"] = dados_limpos["Variável"].replace("PMC - Variação mês/mês imediatamente anterior, com ajuste sazonal (M/M-1)", "Var m/m-1 com A.S")
dados_limpos["Variável"] = dados_limpos["Variável"].replace("PMC - Variação mês/mesmo mês do ano anterior (M/M-12)", "Var m/m-12")
dados_limpos["Variável"] = dados_limpos["Variável"].replace("PMC - Variação acumulada no ano (em relação ao mesmo período do ano anterior)", "Var.Acumulada (ano)")
dados_limpos["Variável"] = dados_limpos["Variável"].replace("PMC - Variação acumulada em 12 meses (em relação ao período anterior de 12 meses)", "Var.Acumulada (12 meses)")
dados_limpos
```

```python
#Visualizar
!pip install plotly
import plotly.express as px
```

```python
#criar gráficos de linhas
fig = px.line(
    data_frame = dados_limpos,
    y = "Valor",
    x = "Data",
    color = "Variável",
    facet_col = "Variável",
    facet_col_wrap = 2,
    title = "Análise PMC"
)

fig.update_yaxes(matches = None, title = "")
fig.update_xaxes(matches = None, title = "")
fig.for_each_annotation(lambda i: i.update(text = i.text.split("=")[-1]))
fig.update_layout(showlegend = False)
fig.add_annotation(
    showarrow = False,
    text = "Mensal",
    x = 0.5,
    xanchor = "center",
    xref= "paper",
    y = 0,
    yanchor = "bottom",
    yref = "paper",
    yshift = -30
)
fig.show()
```

```python
# Restrict data to 2022-2024
dados_filtrados = dados_limpos[
    (dados_limpos['Data'] >= '2022-01') & (dados_limpos['Data'] <= '2024-11')
]

# Create plots using the filtered data
fig = px.line(
    data_frame=dados_filtrados,  # Use filtered data
    y="Valor",
    x="Data",
    color="Variável",
    facet_col="Variável",
    facet_col_wrap=2,
    title="Análise PMC (2022-2024)"  # Update title
)

fig.update_yaxes(matches=None, title="")
fig.update_xaxes(matches=None, title="")
fig.for_each_annotation(lambda i: i.update(text=i.text.split("=")[-1]))
fig.update_layout(showlegend=False)
fig.add_annotation(
    showarrow=False,
    text="Mensal",
    x=0.5,
    xanchor="center",
    xref="paper",
    y=0,
    yanchor="bottom",
    yref="paper",
    yshift=-30,
)
fig.show()
```
